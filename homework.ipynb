```python
import openai
from openai import OpenAI
import os 
```

```python
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

```python
response = client.chat.completions.create(
    model='gpt-3.5-turbo-16k',
    messages=[{"role": "user","content":"is it to late to join the course?" }]
)
```
